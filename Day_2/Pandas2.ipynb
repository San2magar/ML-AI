```python
# sorting by multiple columns
sorted_data_multiple = df.sort_values(by=['sepal_length','sepal_width'],ascending=[False,True])
print('\nSorted data by sepal_length(descending) and sepal_width(ascending)')
print(sorted_data_multiple.head())
```

```python
# filtering the dataset
filtered_data = df[df['petal_length']>1.5]
print('\nFiltered data(petal_length > 1.5):')
(filtered_data.head())
```

```python
# grouping by species and calculating means
grouped_data = df.groupby('species').mean()
print("\nMean Values Grouped by species :")
grouped_data
```

```python
# adding a new column for sepal_area (feature engineering)
df['sepal_area'] = df['sepal_length'] * df['sepal_width']
print("\nDataset with sepal_area column add:")
print(df.head())
```

```python
#eg. function to classify sepal length into categories
def classify_sepal_length(length):
    if length < 5.5:
        return 'Short'
    elif length <6.5:
        return 'Medium'
    else:
        return 'Long'
```

```python
df['sepal_length_category'] = df['sepal_length'].apply(classify_sepal_length)
print("\nDataset with sepal_length_category column add:")
print(df.head())
```

```python
pivot_table = pd.pivot_table(df, values='petal_width', index='species', columns='sepal_length_category', aggfunc='mean')
print("\nDataset with sepal_length_category:")
print(pivot_table)
```

```python
encoded_data = pd.get_dummies(df, columns=['species'])
print("\nDataset with species column encoded:")
print(encoded_data.head())
```

```python
concatenated_data = pd.concat([df, df.copy()], ignore_index=True)
print("\nConcatenated dataset:")
print(concatenated_data.head())
```

```python
melted_data = pd.melt(df,id_vars=['species'],
                     value_vars=['sepal_length','sepal_width','petal_length',
                                'petal_width'])
print("\nMelted dataset:")
print(melted_data.head())
```

```python
missing_columns=[]
for col in data.columns:
    if data[col].isnull().sum()>0:
        missing_columns.append(col)

print(missing_columns)
```

```python
missing_columns={}
for col in data.columns:
    missing_count=data[col].isnull().sum()
    if missing_count > 0:
        missing_percentage=(missing_count/len(data))*100
        missing_columns[col]=round(missing_percentage, 2)

print("\nMissing coulumns with percentage is ")

for col,percentage in missing_columns.items():
    print(f'{col}: {percentage}%')
```
